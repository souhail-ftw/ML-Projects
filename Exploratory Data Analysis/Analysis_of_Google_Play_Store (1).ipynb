````python
def fix_size(s):
  try:
     if s.endswith('M'):
        return float(s[:-1])
     elif s.endswith('k'):
       return float(s[:-1]) / 1024 # convertir en Mb
     elif s == '1,000+':
        return 1.0
     else:
        return np.nan
  except:
      return np.nan
````

````python
fix_size('14M')
````

````python
ds['Size'] = ds['Size'].apply(fix_size)
ds
````
